```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart_disease = pd.read_csv("heart.csv")
```

```python
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins = 10)
```

```python
heart_disease.plot.hist()
```

```python
heart_disease.plot(figsize = (10, 10))
```

```python
heart_disease.plot.hist(figsize = (10, 20), subplots= True);
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
over_50.plot(kind = "scatter",
            x= "age",
            y= "chol",
            c= "target")
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target",
            ax = ax);
ax.set_xlim([55, 75])
```

```python
# Object Oriented method

fig, ax = plt.subplots(figsize = (10, 6))

scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'])
ax.set(title = "Heart Disease and Cholesterol levels",
      xlabel ="Age",
      ylabel = "Cholesterol")

ax.legend(*scatter.legend_elements(), title= "Target");
ax.axhline(over_50["chol"].mean(),
          linestyle = "--")
plt.show()
```

```python
#Plotting two axes using subplots

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                      ncols = 1,
                      figsize = (10, 10),
                        sharex = True)

scatter = ax0.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'])
ax0.set(title = "Heart Disease and Cholesterol levels",
      #xlabel ="Age",
      ylabel = "Cholesterol")

ax0.legend(*scatter.legend_elements(), title= "Target");
ax0.axhline(over_50["chol"].mean(),
          linestyle = "--")

## Plotting on the second axis
scatter = ax1.scatter(x = over_50['age'],
                    y = over_50['thalach'],
                    c = over_50['target'])
ax1.set(title = "Heart Disease and Max Heart rate",
      xlabel ="Age",
      ylabel = "Thalach")

ax1.legend(*scatter.legend_elements(), title= "Target");
ax1.axhline(over_50["thalach"].mean(),
          linestyle = "--")

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
#styling the plot area
plt.style.use("ggplot")

fig, ax = plt.subplots(figsize = (10, 6))

scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'],
                    cmap = "summer")
ax.set(title = "Heart Disease and Cholesterol levels",
      xlabel ="Age",
      ylabel = "Cholesterol")

ax.legend(*scatter.legend_elements(), title= "Target");
ax.axhline(over_50["chol"].mean(),
          linestyle = "--")
plt.show()
```

```python
plt.style.use("ggplot")


fig, (ax0, ax1) = plt.subplots(nrows = 2,
                      ncols = 1,
                      figsize = (10, 10),
                        sharex = True)

scatter = ax0.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'],
                     cmap = "winter")
ax0.set(title = "Heart Disease and Cholesterol levels",
      #xlabel ="Age",
      ylabel = "Cholesterol")
ax0.set_xlim([50, 80])
ax0.set_ylim([60, 500])


ax0.legend(*scatter.legend_elements(), title= "Target");
ax0.axhline(over_50["chol"].mean(),
          linestyle = "--")

## Plotting on the second axis
scatter = ax1.scatter(x = over_50['age'],
                    y = over_50['thalach'],
                    c = over_50['target'],
                     cmap = "winter")
ax1.set(title = "Heart Disease and Max Heart rate",
      xlabel ="Age",
      ylabel = "Thalach")

ax1.legend(*scatter.legend_elements(), title= "Target");
ax1.axhline(over_50["thalach"].mean(),
          linestyle = "--")

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
plt.show()
```
